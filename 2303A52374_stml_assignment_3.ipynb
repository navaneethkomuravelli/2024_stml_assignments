Find out the maximum likelihood parameters values for california_housing [‘latiyude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, each column individually.


```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing(as_frame=True)
df = data.frame[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Latitude']]
# Add a missing column name to match the number of columns in the DataFrame
df.columns = ['median_income', 'house_median_age', 'average_rooms', 'average_bedrooms', 'latitude']

mle_results = {}

for column in df.columns:
    mean_mle = df[column].mean()  # MLE of the mean
    variance_mle = df[column].var()  # MLE of the variance
    mle_results[column] = {'mean': mean_mle, 'variance': variance_mle}

for column, mle in mle_results.items():
    print(f"Column: {column}")
    print(f"  MLE of Mean: {mle['mean']}")
    print(f"  MLE of Variance: {mle['variance']}\n")
```

Find log-likelihood values for 50, 75, and 80. For California_housing [‘latitude’] and find the Max likelihood values.

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
import numpy as np
from scipy.stats import norm

data = fetch_california_housing(as_frame=True)
df = data.frame[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Latitude']]
df.columns = ['median_income', 'house_median_age', 'average_rooms', 'average_bedrooms', 'latitude']

# Calculate MLE for mean and variance of 'latitude'
mean_mle = df['latitude'].mean()
variance_mle = df['latitude'].var()

# Points to evaluate log-likelihood at
points = [50, 75, 80]

for point in points:
    log_likelihood = norm.logpdf(point, loc=mean_mle, scale=np.sqrt(variance_mle)).sum()
    print(f"Log-likelihood at {point}: {log_likelihood}")

print(f"\nMLE of Mean for 'latitude': {mean_mle}")
print(f"MLE of Variance for 'latitude': {variance_mle}")
```

Estimate the density of each column of california_housing [‘latitude’,total_rooms’,’house_median_age’,’total_bedrooms’] data, For the bins [5, 10, 15, 20]. And write the observations.

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
import numpy as np

data = fetch_california_housing(as_frame=True)
df = data.frame[['Latitude', 'AveRooms', 'HouseAge', 'AveBedrms']] # Use the correct column names from the original dataset

bins_list = [5, 10, 15, 20]

for column in df.columns:
    print(f"\nColumn: {column}")
    for bins in bins_list:
        hist, bin_edges = np.histogram(df[column], bins=bins, density=True)

        print(f"  Bins: {bins}")
        for i in range(len(hist)):
            print(f"    Density in bin [{bin_edges[i]:.2f}, {bin_edges[i+1]:.2f}]: {hist[i]:.4f}")

# Observations:
print("\nObservations:")
print("- As the number of bins increases, the density within each bin generally decreases (since the same data is spread across more bins).")
print("- Columns with a wider range of values (like 'total_rooms') tend to have lower densities per bin compared to columns with a narrower range (like 'latitude').")
print("- The distribution of values across bins can give you a sense of the shape of the data (e.g., skewed, uniform, etc.).")
print("- It's important to choose an appropriate number of bins to visualize the density effectively. Too few bins can oversimplify the distribution, while too many can make it difficult to see patterns.")
```
